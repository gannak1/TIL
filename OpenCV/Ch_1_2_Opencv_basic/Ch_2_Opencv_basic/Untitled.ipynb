```python
import numpy as np
import cv2
import sys
import os
import matplotlib.pyplot as plt

img_BGR = cv2.imread('./fig/puppy.bmp',cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


if img is None:
    print('None')
    sys.exit()
    
plt.figure(figsize = (6,4))
plt.subplot(131), plt.imshow(img_BGR), plt.axis('off')
plt.subplot(132), plt.imshow(img_rgb), plt.axis('off')
plt.subplot(133), plt.imshow(img_gray,cmap = 'gray'), plt.axis('off')

plt.show()
```
