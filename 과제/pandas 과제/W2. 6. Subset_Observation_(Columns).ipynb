```python
df.filter(regex = '_')
# _ (under_bar)가 들어간 컬럼만 가지고 옴
```

```python
# df.filter(regex = 'Lengths$')
# length로 끝나는 문자가 사용된 컬러만 가져옴
# 주의) 대소문자 구분
```

```python
df.filter(regex='^sepal')
# Sepal로 시작하는 문자가 사용된 컬럼만 가져옴
```

```python
# df.filter(regex='^x[1-5]$')
# Sepal로 시작하는 문자가 사용된 컬럼만 가져옴
# Matches strings beginning with 'x' and ending with 1,2,3,4,5
```

```python
df.filter(regex='^se')
# se로 시작
```

```python
df.filter(regex = 'es$')
# es로 끝나는 것
```

```python
df.filter(regex ='^(?!species$).*')
# Matches string except the string 'Species'
```

```python
# df.loc[:,'x2':'x4']
# Select all columns between x2 and x4 (inclusive).
```

```python
df.loc[2:4] # 2행에서 4행(포함)까지 가져옴
```
