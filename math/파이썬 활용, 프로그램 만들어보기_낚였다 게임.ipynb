```python
# 게임 플레이 및 사용자 정보 구하기

def game_play(id,pw):
    
    while True:
        df = pd.read_excel('./텍스트게임.xlsx')
        
        돈 = df[(df['ID']==id)&(df['현황']=='판매')]['가격'].sum
        물고기_종류 = len(np.unique(df[df['ID']==id]['물고기']))
        잡은물고기 = df[df['ID']==id]['수량'].sum()
        등급 = len(df[(df['ID']==id)&(df['현황']=='판매')])
        
        if 등급 <= 10:
            등급 = '낚시 준비생'
        elif 등급 > 10 and 등급 <= 20:
            등급 = '낚시 사생팬'
        elif 등급 > 20 :
            등급 = '낚시 고인물'
        
        print('안녕하세요. {} 님 / 레벨 : {} / 가지고 있는 금액 : {} / 잡은 물고기 종류: {} / 잡은 물고기 수량: {}'
        .format(id,등급,돈,물고기_종류,잡은물고기))
        
        선택 = int(input('낚시하기(1) / 삼정이용(2) / 게임종료 (3)'))
        
        if 선택 == 1: # 낚시터
            game_fishing_center(df,id,pw)
        elif 선택 == 2: # 상점
            game_store(df,id,pw)
        elif 선택 == 3 : # 게임종료
            break
```

```python
# 계정정보 

계정정보 = [ ]

df = pd.read_excel("./텍스트게임.xlsx")

계정 = int(input("로그인(1)/회원가입(2)"))

if 계정 == 1: # 로그인 
    계정정보.append(input("ID 입력:"))
    계정정보.append(int(input("패스워드 입력:")))
    
    if 계정정보[0] in df['ID'].unique() and 계정정보[1] in df['PW'].unique():   # unique() :중복제거
        print("게임을 시작합니다.")
        game_play(계정정보[0], 계정정보[1])
    else:
        print("ID 또는 패스워드가 없어요")
        pass

elif 계정 == 2:   # 회원 가입 
    계정정보.append( input("ID 입력"))
    계정정보.append(int(input("패스워드 입력:")))
    
    print("[축하합니다] 회원가입 완료")
    print("게임의 세계로 오실까요?")
    game_play(계정정보[0], 계정정보[1])
```
