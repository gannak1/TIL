#### 주어진 데이터에서 9월의 최고 기온 데이터 시각화 하기

```python
import csv
f = open('seoul_book1.csv')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08' :
            result.append(float(row[-1]))
```

```python
plt.plot(result, 'b^')
plt.show()
```
