```python
song_list = []
songs = soup.select('tbody > tr')
rank =  1
for song in songs:
    title = song.select('a.title')[0].text.strip()
    singer = song.select('a.artist')[0].text.strip()
    mylist = ['지니', rank, title, singer]
    song_list.append(mylist)
    rank += 1
song_list  
```

```python
mydf = pd.DataFrame(song_list, 
                   columns = ['서비스', '순위', '타이틀', '가수'])
mydf.to_excel('./files/genie_class.xlsx', 
             index = False)
```
