```python
# 자동화 프로그램 만들기

import csv

f = open('./age.csv')
data = csv.reader(f)
# data
result = []

name = input('알고 싶은 지역의 이름(읍면동 단위) 입력해 주세요:')

for row in data:
        #print(row)
        if name in row[0]:
            for i in row[3:]:
                k = i.replace(',','')
                result.append(int(k))
print(result)
            
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font',family = 'Malgun Gothic')
plt.title(name + '지역 인구 구조 현황')
plt.plot(result)
plt.show()            
```

```python
import csv

f = open('./gender.csv')
data = csv.reader(f)
# data
m = []
f = []
name = input('알고 싶은 지역의 이름(읍면동 단위) 입력해 주세요:')

for row in data:
    if name in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3].replace(',','')))
            f.append(-int(row[-(i+1)].replace(',','')))
print(m,'\n',f)
f.reverse()
import matplotlib.pyplot as plt

plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()
```

```python
plt.style.use('ggplot')
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.legend() # 범례
plt.title('신도림 지역의 남녀 성별 인구분포')
plt.barh(range(101),m, label = '남자')
plt.barh(range(101),f, label = '여자')
plt.figure(figsize = (10,5), dpi=300)
plt.show()
```
