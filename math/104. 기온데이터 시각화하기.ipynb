```python
import csv
import matplotlib.pyplot as plt

f = open('./seoul_book1.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []
for i in data:
    if i[-1] != '':
        month = i[0].split('-')[1]
        if month == '01':
            jan.append(float(i[-1]))
        if month == '08':
            aug.append(float(i[-1]))


plt.hist([jan,aug],bins = 100, color = ['b','r'])
plt.legend(['aug','jan'])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('./seoul_book1.csv')
data = csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('./seoul_book1.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] =='08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi = 300)
plt.boxplot(day,showfliers=False) # outliers 출력 안함
plt.show()
```
