```python
def get_prod_items(prod_items):
    prod_data = []
    spec_list = []
    for prod_item in prod_items:
        try:
            title = prod_item.select('p.prod_name > a.click_log_product_standard_title_')[0].text
        except:
            title = np.nan
        try:
            sub = prod_item.select('div.spec_list')[0].text.strip('\t')
            if '\t' in sub:
                sub = np.nan
                pass
            else:
                #print(sub)
                pass
        except:
            spec_list.append(np.nan)
        try:
            price = prod_item.select('li.rank_one > p.price_sect > a.click_log_product_standard_price_ > strong')[0].text.strip().replace(',','')
        except:
            price = np.nan
        mylist = [title,sub,price]

        prod_data.append(mylist)
        
    return prod_data
```

```python
def convert_time_minute(time):
    
    try:
        if '시간' in time:
            hour = time.split('시간')[0]
            if '분' in time:
                minute = time.split('시간')[1].replace('분','')
            else:
                minute = 0
        else:
            hour = 0
            minute = time.replace('분','')
        return int(hour)*60 + int(minute)
    except:
        return None
```
