```python
import numpy as np
import pandas as pd
```

```python
sample_1 = pd.read_excel('./files/sample_1.xlsx',header = 1,skipfooter = 2,usecols = 'A:C')
sample_2 = pd.read_excel('./files/sample_2.xlsx',header = 1,skipfooter = 2,usecols = 'A:C',dtype = {'입국객수':np.int64})
code_master = pd.read_excel('./files/sample_codemaster.xlsx')
sample_1
```

```python
sample_1['기준년월'] = '2019-11'
sample_2['기준년월'] = '2019-12'

sample_2.pivot_table(values = '입국객수',index = '성별',columns = '국적코드',aggfunc = 'sum')
```

```python
sample_1_code = pd.merge(left = sample_1,right = code_master, how = 'left',left_on = '국적코드',right_on = '국적코드')
sample_2_code = pd.merge(left = sample_2,right = code_master, how = 'left',left_on = '국적코드',right_on = '국적코드')
```

```python
sample = pd.concat([sample_1_code,sample_2_code])
sample
```
