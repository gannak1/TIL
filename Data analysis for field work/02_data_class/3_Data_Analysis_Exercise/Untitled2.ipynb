```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
driver = webdriver.Chrome(service = Service('../chromedriver_win32/chromedriver.exe'))
url = 'https://www.youtube-rank.com/board/bbs/board.php?bo_table=youtube'
driver.get(url)
soup=BeautifulSoup(driver.page_source,'html.parser')
ranking = soup.select('tr.aos-init')
rank = 1
list_y = []
for i in ranking:
    subject = i.select('td.subject > h1 > p.category')[0].text.split('\n')[0] # 구분
    name = i.select('td.subject > h1 > a')[0].text.split('\t')[-8] # 이름
    subscriber = i.select('td.subscriber_cnt')[0].text
    view = i.select('td.view_cnt')[0].text
    video = i.select('td.video_cnt')[0].text
    list_y.append([subject,rank,name,subscriber,view,video])
youtube = pd.DataFrame(list_y,columns = ['주제','순위','채널명','구독자수','시청자수','비디오수'])
youtube
```

```python
youtube['구독자수'] = youtube['구독자수'].str.replace('만','0000').astype(np.int64)
```

```python
pivot_df = youtube.pivot_table(values = '구독자수',index = '주제',aggfunc = 'sum')
```

```python
pivot_df['subscriber_count']=youtube.value_counts('주제')
```

```python
pivot_df.reset_index(inplace = True)
```
