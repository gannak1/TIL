```python
rank = 1
song_data = []
songs = soup.select('tbody > tr')
for i in songs:
    if len(i.select('th > p.title > a')) > 0:
        song = i.select('th > p.title > a')[0].text
        singer = i.select('td.left > p.artist > a')[0].text
        song_data.append(['bugs',rank,song,singer])
        rank += 1 
```

```python
a = pd.read_excel('./melon_class.xlsx')
b = pd.DataFrame(song_data,columns = ['서비스','순위','타이틀','가수'])
with pd.ExcelWriter('./melon.xlsx') as writer:
    a.to_excel(writer,sheet_name='melon_chart',index = False)
    b.to_excel(writer,sheet_name ='bugs_chart',index = False)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
drive = webdriver.Chrome(service = Service('../chromedriver_win32/chromedriver.exe'))

drive.get('https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220113&hh=14&rtm=Y&pg=1')
soup = BeautifulSoup(drive.page_source,'html.parser')

#genie_clas.xlsx
```
