```python
# -*- coding: utf-8 -*-

### 기본 라이브러리 불러오기
import pandas as pd
import seaborn as sns
```

```python
# load_dataset 함수를 사용하여 데이터프레임으로 변환
df = sns.load_dataset('titanic')
```

```python
# 데이터 탐색
df.info()
```

```python
# NaN값이 많은 deck 열을 삭제, embarked와 내용이 겹치는 embark_town 열을 삭제.

rdf = df.drop(['deck','embark_town'],axis = 1)
rdf
```

```python
rdf.columns.values
```

```python
# age 열에 나이 데이터가 없는 모든 행을 삭제 - age 열(891개 중 177개의 NaN 값)
rdf.dropna(subset = ['age'],how = 'any',axis = 0)
# how : {'any', 'all'}, default 'any'
#     Determine if row or column is removed from DataFrame, when we have
#     at least one NA or all NA.

#     * 'any' : If any NA values are present, drop that row or column.
#     * 'all' : If all values are NA, drop that row or column.
```

```python
# embarked 열의 NaN값을 승선도시 중에서 가장 많이 출현한 값으로 치환하기

rdf.embarked.value_counts()
```

```python
rdf.describe() # 수치형 데이터만 보임
```

```python
most_freq = rdf.describe(include = 'all').embarked.top
```

```python
rdf.embarked.fillna(most_freq, inplace = True)
```

```python
rdf.dropna(subset = ['age'],how = 'any',inplace = True)
```

```python
ndf = rdf[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'embarked']]
```

```python
# 원핫인코딩 - 범주형 데이터를 모형이 인식할 수 있도록 숫자형으로 변환
```

```python
pd.get_dummies(ndf.sex)
```

```python
gender = pd.get_dummies(ndf.sex)
```

```python
pd.concat([ndf,gender],axis = 1)
```

```python
ndf = pd.concat([ndf,gender],axis = 1)
```

```python
onehot_embarked = pd.get_dummies(ndf.embarked,prefix = 'town')
```

```python
ndf = pd.concat([ndf,onehot_embarked],axis = 1)
```

```python
ndf
```

```python
ndf.drop(['embarked','sex'],axis = 1,inplace = True)
ndf
```

```python
ndf.info()
```

```python
# 데이터셋 구분 - 훈련용(train data)/ 검증용(test data)

# 속성(변수) 선택
x=ndf[['pclass', 'age', 'sibsp', 'parch', 'female', 'male', 
       'town_C', 'town_Q', 'town_S']]  #독립 변수 X
y=ndf['survived']                      #종속 변수 Y
```

```python
# 단위가 틀리네? >>> 정규화

from sklearn import preprocessing

x = preprocessing.StandardScaler().fit(x).transform(x)
```

```python
# train data 와 test data로 구분(7:3 비율)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train,x_train.shape,y_train,y_train.shape
```

```python
# KNN 분류 모형

from sklearn.neighbors import KNeighborsClassifier
```

```python
# K = 5 설정

knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# train data로 학습

knn.fit(x_train,y_train)
```

```python
# test data로 y_hat 예측

knn.predict(x_test) # x_test 데이터 입력
```

```python
y_hat = knn.predict(x_test)
```

```python
# 모형 성능 평가

from sklearn import metrics
metrics.confusion_matrix(y_test,y_hat)
```

```python
# 평가 지표

print(metrics.classification_report(y_test,y_hat))
```
