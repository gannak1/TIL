```python
# -*- coding: utf-8 -*-

### 기본 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 준비

pd.read_csv('./data/auto-mpg.csv')
```

```python
pd.read_csv('./data/auto-mpg.csv',header = None)
```

```python
df = pd.read_csv('./data/auto-mpg.csv',header = None)
```

```python
# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
df
```

```python
df.info()
```

```python
df.horsepower.unique()
```

```python
# horsepower 열의 자료형 변경(문자열 -> 숫자)
df.horsepower.replace('?',np.nan,inplace = True) # '?' >> np.nan 변경
df.horsepower.unique()
```

```python
# 누락데이터 행 삭제
df.dropna(subset=['horsepower'],axis = 0,inplace = True)
```

```python
# 데이터 타입 변경(문자형 >> 실수형)
df.horsepower = df.horsepower.astype('float')
```

```python
df.info(), df.shape
```

```python
# 분석에 활용할 열(속성)을 선택 (연비, 실린더, 출력, 중량)
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]
```

```python
# 속성(변수) x(독립변수), y(종속변수)
x = ndf[['cylinders','horsepower','weight']]
y = ndf[['mpg']]
```

```python
# 훈련용, 테스트 데이터 셋 구분

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test, = train_test_split(x,y,test_size = 0.3, random_state = 42)
```

```python
x_train,x_test,x_train.shape,x_test.shape
```

```python
# 다중회귀 분석

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train) # train data로 학습
```

```python
# test data 검증, r2(결정계수)
r2 = lr.score(x_test,y_test)
```

```python
r2
```

```python
# 회귀식의 파라미터(회귀계수, y절편)
lr.coef_ , lr.intercept_
```

```python
# 예측 (x_test로 예측)
lr.predict(x_test)
```

```python
y_hat = lr.predict(x_test)
```

```python
plt.figure(figsize=(10, 5))
ax1 = sns.distplot(y_test, hist=False, label="y_test") # 실제값
ax2 = sns.distplot(y_hat, hist=False, label="y_hat", ax=ax1) # 예측값
plt.show()
```
