```python
# -*- coding: utf-8 -*-

### 기본 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Wholesale customers 데이터셋 가져오기 (출처: UCI ML Repository)
uci_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/\
00292/Wholesale%20customers%20data.csv'
```

```python
pd.read_csv(uci_path, header = 0)
```

```python
df = pd.read_csv(uci_path, header = 0)
```

```python
df.info()
```

```python
df.describe()
```

```python
# 데이터 전처리

x = df.iloc[:,:]
```

```python
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x)
x
```

```python
# K means

from sklearn import cluster
```

```python
# 모델 생성

model = cluster.KMeans(n_clusters = 5, n_init = 10)
```

```python
# 모델 학습

model.fit(x)
```

```python
model.labels_
```

```python
df['cluister'] = model.labels_
```

```python
df
```

```python
# 그래프로 표현 - 시각화
df.plot(kind='scatter', x='Grocery', y='Frozen', c='cluister', cmap='Set1', 
        colorbar=False, figsize=(10, 10))
df.plot(kind='scatter', x='Milk', y='Delicassen', c='cluister', cmap='Set1', 
        colorbar=True, figsize=(10, 10))
plt.show()
```

```python
df.cluister.value_counts()
```

```python
# 큰 값으로 구성된 클러스터 0,1 제외

df[(df.cluister == 0) | (df.cluister == 1)]
```

```python
mask = (df.cluister == 0) | (df.cluister == 1) # 큰 값을 가진 것을 mask로 생성
df[~mask] # 작은 값을 가진 것을 mask로 생성
```

```python
ndf = df[~mask]
```

```python
ndf.plot(kind='scatter', x='Grocery', y='Frozen', c='cluister', cmap='Set1', 
        colorbar=False, figsize=(10, 10))
ndf.plot(kind='scatter', x='Milk', y='Delicassen', c='cluister', cmap='Set1', 
        colorbar=True, figsize=(10, 10))
plt.show()
```

```python
mask = (df.cluister == 0) | (df.cluister == 1) # 큰 값을 가진 것을 mask로 생성
df_mask = df[mask] # 큰 값을 가진 것을 mask로 생성

df_mask.plot(kind='scatter', x='Grocery', y='Frozen', c='cluister', cmap='Set1', 
        colorbar=False, figsize=(10, 10))
df_mask.plot(kind='scatter', x='Milk', y='Delicassen', c='cluister', cmap='Set1', 
        colorbar=True, figsize=(10, 10))
plt.show()
```
