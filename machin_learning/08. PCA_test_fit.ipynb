```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets

np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
y,y.shape
```

```python
df_target = pd.DataFrame(y,columns = ['target'])
df_target
```

```python
import pandas as pd
import numpy as np 

names = ['sepal length','sepal width', 'petal length', 'petal width' ]

df_features = pd.DataFrame(X, columns = names)
df_features
```

```python
from sklearn.preprocessing import StandardScaler

x = df_features.values
y = df_target.loc[:,['target']].values
x,y
```

```python
# 정규화 (비지도학습, fit_transform())

StandardScaler().fit_transform(x)
```

```python
y,y.shape
```

```python
pca = PCA(n_components = 2)

pca_transform = pca.fit_transform(x)
pca_transform
```

```python
pd.DataFrame(data = pca_transform, columns = ['pca_1','pca_2'])
```

```python
df_pca = pd.DataFrame(data = pca_transform, columns = ['pca_1','pca_2'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

pd.concat([df_pca,df_target],axis = 1)
```

```python
df_final = pd.concat([df_pca,df_target],axis = 1)
```

```python
import matplotlib.pyplot as plt
%pylab inline

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('PCA_component_1', fontsize =15)
ax.set_ylabel('PCA_component_2', fontsize =15)
ax.set_title('2 component PCA', fontsize=20)

targets = [0, 1 , 2 ]
colors = ['r','g','b']

for target, color in zip(targets, colors):
    indicesToKeep = df_final['target'] == target
    ax.scatter(df_final.loc[indicesToKeep, 'pca_1'],
               df_final.loc[indicesToKeep, 'pca_2'],
               c = color,
               s = 50)
    
ax.legend(targets)
ax.grid()
```
