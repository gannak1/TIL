```python
# 수치형 자료 변환
# 정규화(normalization)

#1) 정규화 함수 구현


def normal(data):
    min = data.min()
    max = data.max()
    data = (data - min) / (max - min)
    return data


df['']normal(df.Fare)
```

```python
Fare = normal(df.Fare)
Fare
```

```python
# 수치형 자료 변환
# 표준화(Standardzation)

#1) 정규화 함수 구현(standard())

def standard(data):
    data = (data - data.min()) / data.std()
    return data
```

```python
standard(df.Fare)
```
